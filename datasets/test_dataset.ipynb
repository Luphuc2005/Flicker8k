```python

import os
import sys

# Ensure project root is on sys.path for local imports
PROJECT_ROOT = r"D:\HocTap\NCKH_ThayDoNhuTai\Thực nghiệm\Image_Captioning_Flickr8k"
if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)

import torchvision.transforms as transforms
from torch.utils.data import DataLoader

from datasets.preprocessing import load_captions, filter_valid_images, flatten_data
from datasets.vocab import Vocabulary
from datasets.dataset import FlickrDataset
from datasets.collate import collate_fn


# ===== PATH =====
image_dir = r"D:\HocTap\NCKH_ThayDoNhuTai\Thực nghiệm\Image_Captioning_Flickr8k\data\Images"
captions_file = r"D:\HocTap\NCKH_ThayDoNhuTai\Thực nghiệm\Image_Captioning_Flickr8k\data\captions.txt"

# ===== LOAD =====
caps = load_captions(captions_file)
caps = filter_valid_images(image_dir, caps)
image_paths, captions = flatten_data(image_dir, caps)

print("Total samples:", len(image_paths))

# ===== VOCAB =====
vocab = Vocabulary(freq_threshold=5)
vocab.build_vocab(captions)

print("Vocab size:", len(vocab))

# ===== TRANSFORM =====
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

# ===== DATASET =====
dataset = FlickrDataset(image_paths, captions, vocab, transform)

loader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=True,
    collate_fn=collate_fn
)

# ===== TEST =====
for imgs, caps in loader:
    print("Images:", imgs.shape)   # (B, 3, 224, 224)
    print("Captions:", caps.shape) # (B, seq_len)
    break
```
