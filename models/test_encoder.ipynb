```python
# 3. CHƯƠNG TRÌNH TEST (CHẠY TRÊN IPYNB)
# ==========================================

# Cấu hình thử nghiệm
EMBED_DIM = 256
NUM_HEADS = 8
BATCH_SIZE = 4

# Khởi tạo mô hình
cnn_encoder = CNNEncoder(embed_dim=EMBED_DIM)
transformer_encoder = TransformerEncoder(embed_dim=EMBED_DIM, num_heads=NUM_HEADS)

# Tạo dữ liệu giả lập (4 ảnh màu kích thước 224x224)
dummy_input = torch.randn(BATCH_SIZE, 3, 224, 224)

print("--- BẮT ĐẦU KIỂM TRA ---")
print(f"0. Kích thước ảnh đầu vào: {dummy_input.shape}")

# Chạy qua CNN
with torch.no_grad():
    cnn_out = cnn_encoder(dummy_input)
print(f"1. Sau CNN Encoder (Chuỗi hóa): {cnn_out.shape}")
print(f"   -> Giải thích: {BATCH_SIZE} mẫu, mỗi mẫu là chuỗi 49 vector, mỗi vector dài {EMBED_DIM}.")



# Chạy qua Transformer
with torch.no_grad():
    final_out = transformer_encoder(cnn_out)
print(f"2. Sau Transformer Encoder (Ngữ cảnh): {final_out.shape}")
print(f"   -> Giải thích: Kích thước không đổi nhưng các vector đã 'hiểu' lẫn nhau qua Attention.")



print("\n--- TEST THÀNH CÔNG ---")
```
